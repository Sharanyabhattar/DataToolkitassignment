17.   Explain the difference between apply() and map() in Pandas?
-    Map() is used only on Pandas Series and is great for applying a function to each individual value.
For example, you can use map() to convert all values in a column from meters to feet, or to replace values using a dictionary.

🔹 apply() works on both Series and DataFrames, and it's more flexible.
It allows you to apply a function to each row, each column, or each value, depending on how you use it.
For example, you can use apply() to calculate the sum of two columns for each row, or to apply a more complex function that looks at multiple values at once.

So in short:
   map() = simpler, for one column/Series
   apply() = more powerful, for rows, columns, or the whole DataFrame

19. How does Pandas simplify time series analysis?
-   1. DateTime Indexing
Pandas provides powerful date-time indexing that allows you to easily manipulate data based on time.

```python
import pandas as pd

# Create a simple time series
date_range = pd.date_range(start="2023-01-01", periods=5, freq="D")
data = pd.Series([10, 20, 30, 40, 50], index=date_range)

print(data)
```

 Benefits:
The index is now a DatetimeIndex, allowing easy access to data by date.

Slicing works intuitively: data['2023-01-02':'2023-01-04'].

2. Resampling & Frequency Conversion
Pandas allows you to resample data by different time frequencies, like daily to monthly, without any hassle.

```python
# Resample from daily to monthly (mean)
data.resample('M').mean()
```

🔹 Benefits:
Downsampling: Aggregating data to lower frequencies (e.g., from minute to hour).

Upsampling: Filling in missing data when going to a higher frequency.

3. Rolling Windows
Pandas makes it super easy to apply rolling calculations (like moving averages or other window-based functions).

```python
data.rolling(window=3).mean()
```

🔹 Benefits:
Great for calculating things like moving averages, volatility, and smooth trends.

4. Time-Based Indexing and Subsetting
With time series data, you can use date-based slicing directly in the index, making filtering extremely easy.

```python
data['2023-01-03']  # Access data for a specific date
data['2023-01']      # Access data for an entire month
```

🔹 Benefits:
Quickly access specific time periods: days, months, years, or specific ranges.

5. Time Shifting
Pandas allows you to shift data by a certain number of time periods, which is helpful for calculating differences or creating lagged features.

```python
data.shift(1)  # Shifts values by 1 period (creating lag)
```

🔹 Benefits:
Great for calculating day-over-day differences, creating lags for modeling, or shifting data to future/past periods.

6. Handling Missing Data
Time series data often contains missing values. Pandas offers various tools like fillna() and interpolate() to deal with this.




```python
data.fillna(method='ffill')  # Forward fill missing values
```

Pandas makes it easy to handle time zones and convert between them.

```python
data.index = data.index.tz_localize('UTC').tz_convert('US/Eastern')
```

🔹 Benefits:
Work with time series that come from different time zones and convert them to a consistent reference.

Pandas simplifies time series analysis by:

21.  Why is NumPy’s array slicing faster than Python’s list slicing?

- 1.      Contiguous Memory Allocation (C-contiguous arrays)
NumPy arrays are stored in contiguous blocks of memory. This means that elements are stored in a single, compact block. When you slice a NumPy array, it doesn't need to create a copy of the data. Instead, it just creates a view on the original data.

Python lists, on the other hand, are implemented as arrays of pointers, meaning each element is a pointer to an object that could be stored elsewhere in memory. When slicing a list, Python often needs to create a new list and copy each element, which is slower.

2. Efficient C Implementation
NumPy is implemented in C (a lower-level language), and many of its operations, including slicing, are highly optimized at a low level. These optimizations allow slicing to work faster.

Python lists, in contrast, are implemented in Python itself, which is generally slower because it adds a layer of abstraction.

3. No Data Copying (Views vs. Copies)
When slicing a NumPy array, NumPy returns a view (i.e., a reference to the same data in memory) instead of a new object. This means that NumPy doesn’t need to create and store a whole new array, which saves both time and memory.

python
Copy code
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
sliced_arr = arr[1:4]  # View of the original array
Python lists return a new list when sliced, meaning all elements in the slice are copied to a new location in memory.

4. Vectorized Operations and Broadcasting
NumPy arrays support vectorized operations, which allow the application of operations to entire arrays at once, including slicing. This enables faster element-wise operations without the need for Python loops.

python
Copy code
arr[1:4] = [10, 20, 30]  # Fast because it's vectorized
Python lists require more overhead for iteration and modification because they're not optimized for such operations.

5. Memory Views and Strides
NumPy uses strides to determine how to access elements in multi-dimensional arrays. Slicing in NumPy uses stride-based indexing, which is highly efficient because it doesn't require copying data and can use the same memory block but in a different configuration.

Python lists don’t have this system of strides, so they have to handle more overhead when creating slices.

```python
#4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [90, 80, 70, 60, 50],
    'Science': [88, 76, 66, 54, 45],
    'English': [85, 78, 74, 60, 55],
    'History': [70, 65, 60, 58, 50]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title and show the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```
